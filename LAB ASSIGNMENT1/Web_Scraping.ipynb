```python
# ==========================================================
# Name        : Shaza Adelina Binti Khairullah
# Student ID  : SN0107857
# Lab         : Lab Assignment 1 - Web Scraping
# ==========================================================

# Import required libraries
from bs4 import BeautifulSoup
import requests
import csv
import time


# Header used to mimic a real browser request
headers = {
    "User-Agent": "Mozilla/5.0"
}


def scrape_page(url):

    reviews = []

    # Send request to website
    response = requests.get(url, headers=headers)

    # Parse HTML using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Locate all review containers
    review_elements = soup.find_all(
        'div',
        class_='styles_cardWrapper__g8amG styles_show__Z8n7u'
    )

    # Loop through each review
    for review in review_elements:
        try:
            # Extract reviewer name
            user = review.find('span').text.strip()

            # Extract review date
            date = review.find('time')['datetime']

            # Extract review content
            comment = review.find('p').text.strip()

            # Store extracted data
            reviews.append({
                'User': user,
                'Date': date,
                'Comment': comment
            })

        # Skip review if any element is missing
        except:
            continue

    return reviews


# ----------------------------------------------------------
# Scrapes reviews from max of 5 pages
# ----------------------------------------------------------

all_reviews = []

# loop through 5 pages 
for page in range(1, 6):

    print(f"Scraping Page {page}...")

    # Pagination URL
    url = f"https://www.trustpilot.com/review/acerentacar.com?page={page}"

    # Call scraping function
    page_reviews = scrape_page(url)

    # Combine results
    all_reviews.extend(page_reviews)

    # Delay to avoid blocking
    time.sleep(2)


# ----------------------------------------------------------
# to save extracted reviews into CSV file
# ----------------------------------------------------------

with open('ACE_Rent_Car_Review.csv', 'w', newline='', encoding='utf-8') as file:

    fieldnames = ['User', 'Date', 'Comment']

    writer = csv.DictWriter(file, fieldnames=fieldnames)

    # Write column headers
    writer.writeheader()

    # Write review data
    writer.writerows(all_reviews)


print("✅ Reviews successfully saved to ACE_Rent_Car_Review.csv")
```
